```python
import numpy as np

def regra_simpson(f, a, b, n):
    """
    Calcula a integral definida de uma fun√ß√£o 'f' de 'a' at√© 'b'
    usando a Regra de 1/3 de Simpson com 'n' intervalos.

    Par√¢metros:
    f (function): A fun√ß√£o a ser integrada. Deve aceitar um n√∫mero (float/int) e retornar um n√∫mero.
    a (float): O limite inferior de integra√ß√£o.
    b (float): O limite superior de integra√ß√£o.
    n (int): O n√∫mero de intervalos para a aproxima√ß√£o. DEVE SER UM N√öMERO PAR.
    
    Retorna:
    float: O valor aproximado da integral.
    """
    
    # Calcula a largura de cada intervalo (h)
    h = (b - a) / n

    # Soma dos termos nas extremidades
    soma = f(a) + f(b)

    # Soma dos termos com coeficiente 4 (√≠ndices √≠mpares)
    for i in range(1, n, 2):
        x = a + i * h
        soma += 4 * f(x)

    # Soma dos termos com coeficiente 2 (√≠ndices pares)
    for i in range(2, n, 2):
        x = a + i * h
        soma += 2 * f(x)

    # Aplica a f√≥rmula final
    integral = (h / 3) * soma
    return integral

# --- Exemplo de Uso ---

# 1. Defina a fun√ß√£o que voc√™ quer integrar.
#    Vamos usar f(x) = x^2, cuja integral de 0 a 3 √© exatamente 9.
def minha_funcao(x):
    return np.sin(2 * np.pi * x)

# 2. Defina os limites de integra√ß√£o e o n√∫mero de intervalos.
limite_inferior = 0
limite_superior = 1
num_intervalos = 10 # Um n√∫mero par para boa precis√£o

# 3. Chame a fun√ß√£o e imprima o resultado.
resultado = regra_simpson(minha_funcao, limite_inferior, limite_superior, num_intervalos)
print(f"Exemplo 1: Integral de x^2 de {limite_inferior} a {limite_superior}")
print(f"Resultado aproximado: {resultado}")
print("Resultado exato: 0.0\n")

```

```python
######################################################################
#
# Functions to calculate integration points and weights for Gaussian
# quadrature
#
# x,w = gaussxw(N) returns integration points x and integration
#           weights w such that sum_i w[i]*f(x[i]) is the Nth-order
#           Gaussian approximation to the integral int_{-1}^1 f(x) dx
# x,w = gaussxwab(N,a,b) returns integration points and weights
#           mapped to the interval [a,b], so that sum_i w[i]*f(x[i])
#           is the Nth-order Gaussian approximation to the integral
#           int_a^b f(x) dx
#
# This code finds the zeros of the nth Legendre polynomial using
# Newton's method, starting from the approximation given in Abramowitz
# and Stegun 22.16.6.  The Legendre polynomial itself is evaluated
# using the recurrence relation given in Abramowitz and Stegun
# 22.7.10.  The function has been checked against other sources for
# values of N up to 1000.  It is compatible with version 2 and version
# 3 of Python.
#
# Written by Mark Newman <mejn@umich.edu>, June 4, 2011
# You may use, share, or modify this file freely
#
######################################################################

from numpy import ones,copy,cos,tan,pi,linspace

def gaussxw(N):

    # Initial approximation to roots of the Legendre polynomial
    a = linspace(3,4*N-1,N)/(4*N+2)
    x = cos(pi*a+1/(8*N*N*tan(a)))

    # Find roots using Newton's method
    epsilon = 1e-15
    delta = 1.0
    while delta>epsilon:
        p0 = ones(N,float)
        p1 = copy(x)
        for k in range(1,N):
            p0,p1 = p1,((2*k+1)*x*p1-k*p0)/(k+1)
        dp = (N+1)*(p0-x*p1)/(1-x*x)
        dx = p1/dp
        x -= dx
        delta = max(abs(dx))

    # Calculate the weights
    w = 2*(N+1)*(N+1)/(N*N*(1-x*x)*dp*dp)

    return x,w

def gaussxwab(N,a,b):
    x,w = gaussxw(N)
    return 0.5*(b-a)*x+0.5*(b+a),0.5*(b-a)*w
```

```python
import numpy as np
from numpy import *


# Definindo a funcao f(x)-------------------------------------------------
def f(x):
    return x * np.exp(-x)

# Definindo funcao  para executar o metodo do trapezoide------------------
def trapezoide(a,b,n):
	h=(b-a)/n
	s=0
	for i in range(n-1):
		s+=f(a+h*(i+1))
	return h/2*(f(a)+2*s+f(b))

# Definindo funcao  para executar o metodo de Romberg---------------------
def romberg(k,j,a,b):
	if k == 1:
		return trapezoide(a,b,2**(j-1))
	else:
		return (4**(k-1)*romberg(k-1,j+1,a,b)-romberg(k-1,j,a,b))/(4**(k-1)-1)


# Execucao e comparacao dos resultados------------------------------------
ordem = 6
rom = zeros((ordem,ordem))

# Metodo do trapezio------------------------------------------------------
for i in range(0,ordem):
	rom[i,0] = trapezoide(0,11,2**i)

print("Trapezio:\n", rom, "\n\n")

# Metodo de Romberg de maneira analitica----------------------------------
for i in arange(1,ordem):
	for j in arange(i,ordem):
		rom[j,i] = rom[j,i-1]+(rom[j,i-1]-rom[j-1,i-1])/(4**i-1)

print("Romberg:\n",rom, "\n\n")

pontos = []
for i in range(0,6):
	pontos.append(rom[i,0])

plt.plot(pontos)
plt.title("Plot of f(x) = x * np.exp(-x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```
